```python
from tkinter import *
from tkinter import filedialog
import pandas as pd

####Variavel selecionar pasta####

def AbrirArquivo():
    procurar_planilha = filedialog.askopenfilename(initialdir="C:\\Users\\",
                                          title="Qual arquivo?",
                                          filetypes= (("xlsx","*.xlsx"),
                                          ("all files","*.*")))
    planilha = pd.read_excel(procurar_planilha)
    display(planilha)
    
####interface####
    
app= Tk()
####titulo####

app.title("Powerzap")

####tamanho####

app.geometry("500x300")
app.configure(background ="#2F4F4F")

####caminho####

Label(app,text = "Caminho:",background ="#2F4F4F", foreground="#fff", anchor=W ).place(x=10, y=10, width=100, height=20)
vnome = Entry()
vnome.place(x=10, y=30, width=350, height=20)

####botão escolher arquivo####

escolher_arquivo = Button(app, text="Abrir arquivo", command = AbrirArquivo ,anchor=S).place(x=370, y=30, width=120, height=20)

####campo de visualização####

Label(app,text = "Visualização:",background ="#2F4F4F", foreground="#fff", anchor=W ).place(x=10, y=55, width=100, height=20)
vnome2 = Label(app, text="Hello World!")
vnome2.place(x=10, y=75, width=477, height=190)


app.mainloop()
```
